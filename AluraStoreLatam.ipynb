```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda2.head()
def Facturacion_Cat (categoria):
  #Filtra el DF(tienda2) por categoria con el valor de Juguetes y lo guarda en un nuevo DF.
  precioxcat = tienda2[tienda2['Categoría del Producto'] == categoria]
  #Suma el precio y el envió del DF filtrado por Jugietes.
  totprecio = (precioxcat['Precio'].sum()) + (precioxcat['Costo de envío'].sum())
  return totprecio

Tot_Cat = tienda2['Categoría del Producto'].unique()
Indice = 0
for i in Tot_Cat:
  Cat_Actual = i
  TotCat_Actual = Facturacion_Cat(i)
  Indice += 1
  print(f'{Indice}- {Cat_Actual}: {TotCat_Actual}')

```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

conjuntoTiendas = [tienda, tienda2, tienda3, tienda4]
TotxTienda = []
for i in conjuntoTiendas:
  Total = i['Precio'].sum()
  TotxTienda.append(Total)
datos = {
    "Tiendas": ["Tienda_1", "Tienda_2", "Tienda_3", "Tienda_4"],
    "Subtotal": TotxTienda
}
dfTiendas = pd.DataFrame(datos)
print(dfTiendas)
```

# 3. Calificación promedio de la tienda


```python
TiendasConjunto = [tienda, tienda2, tienda3, tienda4]
numero = 1

for i in TiendasConjunto:
  notaXcat = i[["Categoría del Producto", "Calificación"]].groupby("Categoría del Producto").mean()
  notaXtienda = i["Calificación"].mean()
  redondeo1 = round(notaXcat, 2)
  redondeo2 = round(notaXtienda, 2)
  print(f"Tienda {numero}")
  print(redondeo1)
  print(f"Calificación final:            {redondeo2}/5")
  print("-------------------------------------")
  numero += 1
```

```python
#Grafico de promedio
#Variables
mediaXtienda = []
numTienStr = []
num = 1

#Añede la calificación promedio por tienda en una lista.
for i in TiendasConjunto:
  mediaXtienda.append(i["Calificación"].mean())
  numTienStr.append(f"Tienda:{num}")
  num += 1

#Crea un nuevo DF para el promedio de las tiendas.
mediaTienda = pd.DataFrame({
    "Tiendas": numTienStr,
    "Promedio": mediaXtienda
})

#Se definen las dimensiones del grafico.
plt.figure(figsize=(5, 2))
sns.set(style="ticks")
sns.scatterplot(x="Tiendas", y="Promedio", data=mediaTienda, palette="pastel")
sns.despine()
plt.title("Calificación por Tienda")
plt.show()
```

# 4. Productos más y menos vendidos

```python
from tabulate import tabulate
num1 = 0

def crealista (SimLista, ValorList):
  SimLista.append(ValorList)

def creaGrafico (ejex, ejey, tiendaIterada,titulo):
  plt.figure(figsize=(5,2))
  sns.set_style("darkgrid")
  sns.barplot(x=ejex, y=ejey, data=tiendaIterada, palette="pastel")
  plt.title(titulo)
  plt.show()

for i in TiendasConjunto:
  conteoProductos = i["Producto"].value_counts()
  conteo = round((conteoProductos/2359)*100)
  cantidad = i["Producto"].unique()
  listProd = []
  listContProd = []
  listTotalProd = []
  num1 += 1

  for i in cantidad:
    crealista(listProd, i)
  for i in conteo:
    crealista(listContProd, i)
  for i in conteoProductos:
    crealista(listTotalProd, i)

  producMaxMiN = pd.DataFrame({
      "Producto": listProd,
      "Porcentaje": listContProd,
      "Total": listTotalProd
  })

  VentasMax = producMaxMiN.nlargest(3, "Total")
  VentasMin = producMaxMiN.nsmallest(3, "Total")
  print(f"Tienda:{num1} Productos con mayores ventas.")
  print(tabulate(VentasMax, headers="keys", tablefmt="fancy_grid"))
  creaGrafico(VentasMax['Total'],VentasMax['Producto'], VentasMax, f"Tienda: {num1}")
  print(f"Tienda:{num1} Productos con menores ventas.")
  print(tabulate(VentasMin, headers="keys", tablefmt="fancy_grid"))
  creaGrafico(VentasMin['Total'],VentasMin['Producto'], VentasMin, f"Tienda: {num1}")
```

# 5. Envío promedio por tienda

```python
conjuntoTiendas = [tienda, tienda2, tienda3, tienda4]
numero = 1
for i in conjuntoTiendas:
  promTienda = i["Costo de envío"].mean()
  redondear = round(promTienda, 2)
  print(f"Tienda {numero}. Costo de envió promedio: {redondear}")
  numero += 1
  print("----------------------------------")
```

```python
#Variables.
numero = 1
varCuan = []
varCual = []

#El bucle agrega datos a las listas.
for i in conjuntoTiendas:
  varCuan.append(i['Costo de envío'].mean())
  varCual.append(f"Tienda: {numero}")
  numero += 1

#Crea un DataFrame.
costPromEnvio = pd.DataFrame({
    "Costo": varCuan,
    "Tienda": varCual
})

#Crea el grafico comparativo de costos de envió promedio por tienda.
plt.figure(figsize=(5, 2))
sns.set_style("dark")
sns.barplot(x="Costo", y="Tienda", data=costPromEnvio, palette="pastel")
plt.title("Costo promedio de envío por Tienda")
plt.show()
```
